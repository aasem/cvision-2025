**Just a quick implementation of Canny Edge Detector**

```python
import cv2
import os
import numpy as np
from matplotlib import pyplot as plt
```

```python
!git clone https://github.com/aasem/cvision2025.git
```

```python
image_path = '/content/cvision2025/data/cat.png'
img = cv2.imread(image_path)
plt.imshow(img)
np.shape(img)
```

```python
edges = cv2.Canny(img,100,200)
plt.imshow(edges)
np.shape(img)
```

```python
plt.subplot(121), plt.imshow(img)
plt.title('Original Image'), plt.xticks([]), plt.yticks([])
plt.subplot(122), plt.imshow(edges)
plt.title('Edge Image'), plt.xticks([]), plt.yticks([])
```
